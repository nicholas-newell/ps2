```python
import pandas as pd

df = pd.read_csv('/blue/zoo6927/share/Jupyter_Content/data/auto_mpg.csv')
```

```python
df
```

```python
# mpg vs. weight
x = df[['weight']]
y = df[['mpg']]
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression().fit(x,y)
```

```python
print(model.intercept_, model.coef_)
```

```python
model.score(x,y)
```

```python
df['weight_sq']= df.weight**2
x2= df[['weight', 'weight_sq']]
```

```python
model = LinearRegression().fit(x2,y)
print(model.intercept_, model.coef_)
```

```python
model.score(x2,y)
```

```python
import numpy as np
import pandas as pd
import plotnine as pn
xvals = np.linspace(x2.weight.min(), x2.weight.max(), len(df))
x_predict = pd.DataFrame({
    'weight': xvals,
    'weight_sq': xvals**2
})
```

```python
y_predict = model.predict(x_predict)
```

```python
pn.ggplot(df, pn.aes(x='weight', y='mpg')) + pn.geom_point() + pn.geom_line(mapping=pn.aes(x=xvals, y=y_predict), color='red', size=1.0)
```
